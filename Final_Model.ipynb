For testing data from XML formatted .sgm files, please use test_xml_sgm().
<br><br>
It takes four parameters:  
<br>
- **folder_path:** This should be a string that contains the path where the test .sgm files are saved.
- **model:** This should be the trained model. Please use model=nb_model.
- **vectorizer:** This should be the word vectors that were previously calculated. Please use vectorizer=vectorizer.
- **drop_na:** This should either be True of False. When it is equal to True, any data that is missing a topic or body text, is removed before predictions are made. The default is True.
<br><br>
This function is meant to work only with files that are similar to the ones provided. It will look in the folder specified for any files with a .sgm extension and try to parse them.
